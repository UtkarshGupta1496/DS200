```
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
WS = pd.read_excel('Length of National Highways.xlsx')
WS_np = np.array(WS)
```

```
labels = [WS_np[7][0],WS_np[12][0], WS_np[2][0],WS_np[3][0],WS_np[4][0]]
```

```
year_2014 = [WS_np[7][11],WS_np[12][11], WS_np[2][11],WS_np[3][11],WS_np[4][11]]
year_2015 = [WS_np[7][12],WS_np[12][12], WS_np[2][12],WS_np[3][12],WS_np[4][12]]
year_2013 = [WS_np[7][10],WS_np[12][10], WS_np[2][10],WS_np[3][10],WS_np[4][10]]
```

```
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, year_2014, width, label='2014')
rects2 = ax.bar(x + width/2, year_2015, width, label='2015')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length of Highways',fontweight='bold')
ax.set_xlabel('States',fontweight='bold')
ax.set_title('Length of Highways for Year 2014 and 2015',fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
plt.show()
```

```
# Box Plots

# 2014 Data
spread_2014 = np.array(year_2014)

# 2015 Data
spread_2015 = np.array(year_2015)

# 2013 Data
spread_2013 = np.array(year_2013)

fig1, ax1 = plt.subplots()
ax1.set_ylabel('  Length of highways \n across various states',fontweight='bold')
ax1.set_xlabel('Year',fontweight='bold')
ax1.boxplot([spread_2013,spread_2014, spread_2015])
plt.xticks([1, 2, 3], ['2013','2014','2015'])
```

```
# Scatter Plots

fig, ax = plt.subplots()

scatter = ax.scatter(labels, year_2014)

plt.ylabel(' Length of Highways(2014)',fontweight='bold')
plt.xlabel('States',fontweight='bold')

ax.grid(True)
```
